## Helpful Tips / Practical Points

### Debugging

There are multiple ways to debug code, and here are just a few approaches you can take:
- Break the problem down into small parts, and evaluate whether or not each one is correct.
- Print out input / output to see if what you have is expected
- Debug on a simple case that is easy to understand and runs fast. This way you can more easily spot check if everything is correct, and you do not have to wait a long time to see if your fix helped.
